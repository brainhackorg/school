# Figure plotting

### Description
This python code will generate and save the figure of the plot that I used in my final report.
### Format
In order for this code to code please follow the same formart I used in the .CSV file.

Note: Make sure that the .CSV file and this file are in the same directory in order for it to work properly.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the Excel file
df = pd.read_csv('susceptibility.csv')
```

```python
# Set the figure size and spacing
plt.figure(figsize=(10, 8))
plt.subplots_adjust(hspace=0.6, wspace=0.3)

# Plot the first set of figures (Challenge protocol)
plt.subplot(2, 1, 1)
plt.title('Variability Assessment of Brain Region Values', fontsize=16, pad=50)
sns.regplot(x='Chi', y='QSM', data=df, label='$\chi^{\t{Total}}_{\t{Simulated}}$ vs QSM',scatter=True,ci=None)
sns.regplot(x='Chi', y='ChiTotal', data=df, label='$\chi^{\t{Total}}_{\t{Simulated}}$ vs $\chi^{\t{Total}}_{\t{Measured}}$', scatter=True,ci=None,color='orange')
plt.plot([df['Chi'].min(), df['Chi'].max()], [df['Chi'].min(), df['Chi'].max()], color='red', alpha=0.5,label='y=x',linestyle='dashed')
plt.xlabel('$\chi^{\t{Total}}_{\t{Simulated}}$ in (ppm)')
plt.ylabel('QSM and $\chi^{\t{Total}}_{\t{Measured}}$  in (ppm)')
plt.legend()
plt.text(0.05, 0.12, "R\u00B2=0.98 (slope=0.77)", fontsize=12, ha='center',color='blue')
plt.text(0.05, 0.1, "R\u00B2=0.98 (slope=0.39)", fontsize=12, ha='center',color='orange')


plt.subplot(2, 2, 3)
sns.regplot(x='ChiPos_truth', y='ChiPosMap', data=df, label='$\chi^{\t{Postive}}_{\t{Simulated}}$ vs $\chi^{\t{Positive}}_{\t{Measured}}$ in (ppm)', scatter=True,ci=None, color='green')
plt.plot([df['ChiPos_truth'].min(), df['ChiPos_truth'].max()], [df['ChiPos_truth'].min(), df['ChiPos_truth'].max()], color='red', alpha=0.5,label='y=x',linestyle='dashed')
plt.xlabel('$\chi^{\t{Postive}}_{\t{Simulated}}$ in (ppm)')
plt.ylabel('$\chi^{\t{Positive}}_{\t{Measured}}$ in (ppm)')
plt.legend()
plt.text(0.05, 0.1, "R\u00B2=0.99 (slope=0.36)", fontsize=12, ha='center',color='green')

plt.subplot(2, 2, 4)
sns.regplot(x='ChiNeg_truth', y='ChiNegMap', data=df, label='$\chi^{\t{Negative}}_{\t{Simulated}}$ vs $\chi^{\t{Negative}}_{\t{Measured}}$', scatter=True,ci=None, color='purple')
plt.plot([df['ChiNeg_truth'].min(), df['ChiNeg_truth'].max()], [df['ChiNeg_truth'].min(), df['ChiNeg_truth'].max()], color='red', alpha=0.5,label='y=x',linestyle='dashed')
plt.xlabel('$\chi^{\t{Negative}}_{\t{Simulated}}$ in (ppm)')
plt.ylabel('$\chi^{\t{Negative}}_{\t{Measured}}$ in (ppm)')
plt.legend()
plt.text(-0.02, -0.01, "R\u00B2=0.53 (slope=0.15)", fontsize=12, ha='center',color='purple')

# Save the figure
plt.savefig('comparison with variation.png')

# Show the plots
plt.show()
```
